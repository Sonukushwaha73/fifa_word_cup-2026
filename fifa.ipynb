# --------------->Visualization (Matplotlib / Seaborn)<------------------------

# Position count ka bar chart

```python
import matplotlib.pyplot as plt

# Position-wise player count
position_count = df["position"].value_counts()

plt.figure(figsize=(8, 5))
plt.bar(position_count.index, position_count.values)

plt.title("Position Count")
plt.xlabel("Position")
plt.ylabel("Number of Players")

plt.show()
```

# Top 10 nationalities ka bar chart

```python
import matplotlib.pyplot as plt

# Top 10 nationalities
top10 = df["nationality"].value_counts().head(10)

plt.figure(figsize=(10, 6))
plt.bar(top10.index, top10.values)

plt.title("Top 10 Nationalities by Player Count")
plt.xlabel("Nationality")
plt.ylabel("Number of Players")
plt.xticks(rotation=45)

plt.show()
```

# Age distribution ka histogram.

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8, 5))

plt.hist(df["age"], bins=10)

plt.title("Age Distribution")
plt.xlabel("Age")
plt.ylabel("Frequency")

plt.show()
```

# Age ka boxplot

```python
plt.figure(figsize=(6, 5))

plt.boxplot(df["age"])

plt.title("Boxplot of Age")
plt.ylabel("Age")

plt.show()
```

# Team-wise average age ka bar chart.

```python


Position vs Age boxplot.
Nationality vs Player Count countplot.
Age ka KDE plot.
Correlation heatmap (numeric columns).
Pairplot (numeric columns).
```

```python
team_avg_age = df.groupby("team")["age"].mean()

plt.figure(figsize=(10, 6))
plt.scatter(team_avg_age.index, team_avg_age.values)

plt.title("Team-wise Average Age")
plt.xlabel("Team")
plt.ylabel("Average Age")

plt.xticks(rotation=45)
plt.show()
```

# Position-wise average age ka bar chart.

```python
# Position-wise average age
position_avg_age = df.groupby("position")["age"].mean()

plt.figure(figsize=(8, 5))
plt.bar(position_avg_age.index, position_avg_age.values)

plt.title("Position-wise Average Age")
plt.xlabel("Position")
plt.ylabel("Average Age")

plt.xticks(rotation=45)
plt.show()
```

# Position vs Age boxplot.

```python
plt.figure(figsize=(10, 6))

df.boxplot(column="age", by="position")

plt.title("Position vs Age")
plt.suptitle("") 
plt.xlabel("Position")
plt.ylabel("Age")

plt.show()
```

# Nationality vs Player Count countplot

```python
plt.figure(figsize=(12, 6))

sns.countplot(
    data=df,
    x="nationality",
    order=df["nationality"].value_counts().index
)

plt.title("Nationality vs Player Count")
plt.xlabel("Nationality")
plt.ylabel("Player Count")
plt.xticks(rotation=90)

plt.show()
```

# Age ka KDE plot.

```python
plt.figure(figsize=(8, 5))

sns.kdeplot(data=df, x="age", fill=True)

plt.title("Age KDE Plot")
plt.xlabel("Age")
plt.ylabel("Density")

plt.show()
```

```python
numeric_df = df.select_dtypes(include=["number"])

# Pairplot
sns.pairplot(numeric_df)

plt.show()
```
